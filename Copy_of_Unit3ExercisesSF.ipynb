## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task9**:

Plot the pdfs (techincally the pmfs) of the following distributions: $Bernoulli(p=0.7)$, $Binomial(n=10,p=0.7)$, $Poisson(λ=7)$, $Geometric(p=0.7)$, $HyperGeometric(N=50,k=35,n=15)$, $Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4])$

Additionally, compute the first four moments for all those distributions--except the Categorical distribution.

````python
pz.distributions.bernoulli.Bernoulli(p=0.7).plot_pdf(support=(0,3))
a=pz.distributions.bernoulli.Bernoulli(p=0.7)
pz.distributions.binomial.Binomial(n=10, p=0.7).plot_pdf(support=(0,3))
b=pz.distributions.binomial.Binomial(n=10, p=0.7)
pz.distributions.poisson.Poisson(mu=7).plot_pdf(support=(0,3))
c=pz.distributions.poisson.Poisson(mu=7)
pz.distributions.geometric.Geometric(p=0.7).plot_pdf(support=(0,3))
d=pz.distributions.geometric.Geometric(p=0.7)
pz.distributions.hypergeometric.HyperGeometric(N=50,k=35,n=15).plot_pdf(support=(0,3))
e=pz.distributions.hypergeometric.HyperGeometric(N=50,k=35,n=15)
pz.distributions.categorical.Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4]).plot_pdf()
````

````python
v=a.var(),a.skewness(),a.kurtosis(),a.mean()

n=b.var(),b.skewness(),b.kurtosis(),b.mean()

m=c.var(),c.skewness(),c.kurtosis(),c.mean()

p=d.var(),d.skewness(),d.kurtosis(),d.mean()

u=a.var(),e.skewness(),e.kurtosis(),e.mean()
v,n,m,p,u
````
