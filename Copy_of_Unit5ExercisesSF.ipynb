**Task5**:

Create and run a stright line linear regression model on the howell data set with height predicting weight. Visualize your model and its posterior.

```python

with pm.Model() as bm_hmodel:
    #priors
     = pm.Normal("", mu=0, sigma=3)
     = pm.Normal("", mu=0, sigma=10)
     = pm.HalfNormal("", 1)
    m = pm.Deterministic('m', +  * howell.height)

    #likelihood
    #the linear part of the model is put directly into mu, I just stole all this code from the notes
    y = pm.Normal("y_pred",
                  mu=m,
                  sigma=,
                  observed=howell['weight'])
    ibm_hdata = pm.sample()
```

```python
az.rhat(ibm_hdata)
az.plot_trace(ibm_hdata)

az.plot_posterior(ibm_hdata)
pm.sample_posterior_predictive(ibm_hdata, model=bm_hmodel, extend_inferencedata=True)
az.plot_ppc(ibm_hdata, num_pp_samples=100, figsize=(12, 4), colors=["C1", "C0", "C1"])
```

```python
posterior = az.extract(ibm_hdata, num_samples=100)

# grabbing x values for graphing.
x_plot = xr.DataArray(
    np.linspace(howell.height.min(), howell.height.max(), 60),
    dims="plot_id"
    )

# this creates the expected line, the path we predict temperature and
#rented bike count deviates from
mean_line = posterior[""].mean() + posterior[""].mean() * x_plot

#These are a 50 lines that our model came up with
lines = posterior[""] + posterior[""] * x_plot

#set up plot
_, ax = plt.subplots()

#plots 50 other lines our model came up with.
ax.plot(x_plot, lines.T, c="C1", alpha=0.2, label="lines")

#plots the mean line
ax.plot(x_plot, mean_line, c="C0", label="mean line")

#plot the raw data
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)

#label axes and create legend
ax.set_xlabel("height")
ax.set_ylabel("weight")
```

It looks more like an exponential graph and that the y-intercept is way higher than orignal thought.
