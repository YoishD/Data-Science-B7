**Task5**:

Create and run a stright line linear regression model on the howell data set with height predicting weight. Visualize your model and its posterior.

```python

with pm.Model() as bm_hmodel:
    #priors
    α = pm.Normal("α", mu=0, sigma=3)
    β = pm.Normal("β", mu=0, sigma=10)
    σ = pm.HalfNormal("σ", 1)
    m = pm.Deterministic('m',α + β * howell.height.values)

    #likelihood
    #the linear part of the model is put directly into mu, I just stole all this code from the notes
    y = pm.Normal("y_pred",
                  mu=m,
                  sigma=σ,
                  observed=howell['weight'])
    ibm_hdata = pm.sample()
```

```python
az.rhat(ibm_hdata)
az.plot_trace(ibm_hdata)

az.plot_posterior(ibm_hdata)
pm.sample_posterior_predictive(ibm_hdata, model=bm_hmodel, extend_inferencedata=True)
az.plot_ppc(ibm_hdata, num_pp_samples=100, figsize=(12, 4), colors=["C1", "C0", "C1"])
```

```python
posterior = az.extract(ibm_hdata, num_samples=100)

# grabbing x values for graphing.
x_plot = xr.DataArray(
    np.linspace(howell.height.min(), howell.height.max(), 60),
    dims="plot_id"
    )

# this creates the expected line, the path we predict temperature and
#rented bike count deviates from
mean_line = posterior["α"].mean() + posterior["β"].mean() * x_plot

#These are a 50 lines that our model came up with
lines = posterior["α"] + posterior["β"] * x_plot

#set up plot
_, ax = plt.subplots()

#plots 50 other lines our model came up with.
ax.plot(x_plot, lines.T, c="C1", alpha=0.2, label="lines")

#plots the mean line
ax.plot(x_plot, mean_line, c="C0", label="mean line")

#plot the raw data
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)

#label axes and create legend
ax.set_xlabel("height")
ax.set_ylabel("weight")
```

It looks more like an exponential graph and that the y-intercept is way higher than orignal thought.

**Task7**:

Create and visualize the posterior of a generalized linear model that improves the on the fit of the model you made in Task5.

```python
with pm.Model() as bm_hmodel:

    α = pm.Normal("α", mu=0, sigma=1)
    β = pm.Normal("β", mu=0, sigma=10)

    #the linear part of our model,
    #but with a twist:
    #our line is exponentiated, in order to make our all our values positive
    m = pm.Deterministic("m", pm.math.exp(α + β * howell.height))

    #prior for the likelihood's standard deviation
    σ = pm.HalfNormal("σ", 10)

    #likelihood
    y_pred = pm.NegativeBinomial("y_pred", mu=m, alpha=σ, observed=howell.weight)

    # Sampling
    ibm_hdata = pm.sample(target_accept=0.99)
```

```python
az.rhat(ibm_hdata)
az.plot_trace(ibm_hdata)
```

```python
#grabs 100 posterior samples
posterior = az.extract(ibm_hdata, num_samples=100)

# grabbing x values for graphing.
x_plot = xr.DataArray(
    np.linspace(howell.height.min(), howell.height.max(), 50),
    dims="plot_id"
    )

# this creates the expected line, the path we predict temperature and
#rented bike count deviates from
mean_line = np.exp(posterior["α"].mean() + posterior["β"].mean() * x_plot)

#These are a 50 lines that our model came up with
lines = np.exp(posterior["α"] + posterior["β"] * x_plot)

#set up plot
_, ax = plt.subplots()

#plots 50 other lines our model came up with.
ax.plot(x_plot, lines.T, c="C1", alpha=0.2, label="lines")

#plots the mean line
ax.plot(x_plot, mean_line, c="C0", label="mean line")

#plot the raw data
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)

#label axes and create legend
ax.set_xlabel("height")
ax.set_ylabel("weight")
```

```python
pm.sample_posterior_predictive(ibm_hdata, model=bm_hmodel, extend_inferencedata=True)

mean_line = ibm_hdata.posterior["m"].mean(("chain", "draw"))
height = np.random.normal(howell.height.values, 0.01)
idx = np.argsort(height)
x = np.linspace(height.min(), height.max(), 15)
y_pred_q = ibm_hdata.posterior_predictive["y_pred"].quantile(
    [0.03, 0.97, 0.25, 0.75], dim=["chain", "draw"]
)
y_hat_bounds = iter(
    [
        PchipInterpolator(height[idx], y_pred_q[i][idx])(x)
        for i in range(4)
    ]
)

_, ax = plt.subplots()
ax.plot(howell.height, howell.weight, "C2.", zorder=-3)
ax.plot(howell.height[idx], mean_line[idx], c="C0")

for lb, ub in zip(y_hat_bounds, y_hat_bounds):
    ax.fill_between(x, lb, ub, color="C1", alpha=0.5)

ax.set_xlabel("height")
ax.set_ylabel("weight")
```
